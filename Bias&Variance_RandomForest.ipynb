```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
np.random.seed(42)
X, y = make_circles(n_samples=300, noise=0.4, factor=0.2, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X,y)

def decision_boundary(clf):
  plt.figure(figsize=(9,6))
  x_range = np.linspace(X.min(), X.max(), 500)
  xx1, xx2 = np.meshgrid(x_range,x_range)
  input_array = np.array([xx1.ravel(),xx2.ravel()]).T

  y_pred = clf.predict(input_array)
  plt.contourf(xx1,xx2,y_pred.reshape(xx1.shape),alpha=0.45)
  plt.scatter(X[:,0],X[:,1],c=y)

decision_boundary(dtree)
```

```python
# let's observe same stuff with random forest

from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X,y)

decision_boundary(rf)
```
